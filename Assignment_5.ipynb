Question-1: Principal Component Analysis


```python
#Reading the CC_GENERAL.csv file
df = pd.read_csv("/content/CC GENERAL.csv")
df.head()
```

```python
#checking if there is any null data in the dataset using isnull() function
df.isnull().any()
```

```python
#replacing the null data with the mean
df.fillna(df.mean(), inplace=True)
df.isnull().any()
```

```python
x = df.iloc[:,1:-1]
y = df.iloc[:,-1]
print(x.shape,y.shape)
```

```python
# a. Apply PCA on CC dataset.
#PCA is used to analyze the datasets inorder to remove the redundant features without losing too much information
pca = PCA(3)
x_pca = pca.fit_transform(x)
principalDf = pd.DataFrame(data = x_pca, columns = ['principal component 1', 'principal component 2', 'principal component 3'])
finalDf = pd.concat([principalDf, df.iloc[:,-1]], axis = 1)
finalDf.head()
```

```python
# b. Apply k-means algorithm on the PCA result and report your observation if the silhouette score has improved or not?

X = finalDf.iloc[:,0:-1] #predictor variable containing all rows of finalDf except for the last column, representing the principal components generated by PCA
y = finalDf.iloc[:,-1] #target variable containing only the last column of finalDf, representing the target variable
print(X.shape,y.shape)
```

```python
nclusters = 3 
km = KMeans(n_clusters=nclusters)
km.fit(X)

# predict the cluster for each data point
y_cluster_kmeans = km.predict(X)

# Summary of the predictions made by the classifier
print(classification_report(y, y_cluster_kmeans, zero_division=1))
print(confusion_matrix(y, y_cluster_kmeans))

#finding the accuracy
train_accuracy = accuracy_score(y, y_cluster_kmeans)
print("\nAccuracy for our Training dataset with PCA:", train_accuracy)

#Calculating sihouette Score
score = metrics.silhouette_score(X, y_cluster_kmeans)
print("Sihouette Score: ",score)

```

Question-4: Briefly identify the difference between PCA and LDA



PCA (Principal Component Analysis) and LDA (Linear Discriminant Analysis) are two common dimensionality reduction approaches in machine learning. They do, however, serve various aims and employ different methods:

Purpose:
PCA is an unsupervised learning technique that identifies the directions of maximum variance in a dataset. It minimizes the number of features by converting the original dataset into a new coordinate system with uncorrelated features sorted by variance. PCA is a popular technique for data compression, visualization, and noise reduction.
LDA, on the other hand, is used for supervised learning and seeks the linear combinations of characteristics that best distinguish between classes. It minimizes the amount of features by projecting the original dataset onto a lower-dimensional space and maximizes class separability. LDA is often used to extract features, recognize patterns, and classify data.

Method:
PCA operates by determining the eigenvectors and eigenvalues of the data's covariance matrix. The eigenvectors represent the highest variance directions, whereas the eigenvalues represent the amount of variation explained by each eigenvector. The top k eigenvectors are chosen by PCA, where k is the desired dimensionality of the reduced dataset.
LDA, on the other hand, increases the data's between-class scatter while minimizing its within-class scatter. The eigenvectors and eigenvalues of the product of two matrices: the between-class scatter matrix and the within-class scatter matrix, must be determined. LDA chooses the top k eigenvectors with the highest eigenvalues.
